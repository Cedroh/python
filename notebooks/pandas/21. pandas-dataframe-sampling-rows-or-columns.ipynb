# 21. Pandas - Random Sample of a subset of a dataframe

```python
import pandas as pd
df = pd.read_csv("../csv/movie_metadata.csv")
```

## 1. Random sampling of rows, columns from DataFrame with sample()

```python
# Default behavior of sample()
df.sample()
```

```python
# return n rows
df.sample(3)
```

```python
# columns
df.sample(3, axis=1).head()
```

```python
# The fraction of rows and columns: frac
df.sample(frac=0.001)
```

```python
# sample with seed
df.sample(n=3, random_state=5)
```

## 2. p.random.choice

```python
# nu,py choice for DataFrame sampling
import numpy as np
chosen_idx = np.random.choice(1000, replace=False, size=5)
df.iloc[chosen_idx]
```

## 3. Random sample of rows based on column values

```python
# conditional DataFrame sampling few values - separate display 
col = 'color'
for typ in list(df[col].dropna().unique()):
    print(typ, end=' - ')
    display(df[df[col] == typ].sample(3))
```

```python
# conditional DataFrame sampling many values - grouped display 
col = 'content_rating'
sample = []

variants = list(df[col].dropna().unique())
print(variants)

for typ in variants:
    sample.append(df[df[col] == typ].sample())
pd.concat(sample)
```

## 4. Dataframe sampling with numpy and weights

```python
# excluding 'Color' values by applying weights 0 - Color and 1 - rest
df['weights'] = np.where(df['color'] == 'Color', .0, 1)
df.sample(frac=.001, weights='weights')
```

```python
# Including only 'Color' values by applying weights 1 - Color and 0 - rest
df['weights'] = np.where(df['color'] == 'Color', 1, 0.0)
df.sample(frac=.001, weights='weights')
```

```python
np.where(df['color'] == 'Color', 1, 0)[270:280]
```

```python
print(list(df['color'][270:280]))
```

```python
# Including/excluding list of values - with equal probability
df['weights'] = np.where(df['content_rating'].isin(['PG-13', 'PG', 'G']), 1, 0)
df.sample(frac=.002, weights='weights')['content_rating']
```

```python
df[df['content_rating'].isin(['PG-13', 'PG', 'G'])]['content_rating'].value_counts()
```

## 5. Pandas sample rows by group

```python
df.groupby('color').apply(lambda x: x.sample(n=3))
```

```python
df.groupby('color').apply(lambda x: x.sample(n=3)).reset_index(drop = True)
```
